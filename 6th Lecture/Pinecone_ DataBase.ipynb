# Generate the Embedding :

- To generate the embedding, also I am going to use the same function, previous one.

```python
import requests
import numpy as np

def generate_embeddings(text):
    url = "https://api.euron.one/api/v1/euri/embeddings"
    headers = {
        "Content-Type": "application/json",
        "Authorization": "Bearer euri-46d7971d2ffa9e1c356959de86118e7c51ab8390c3d8608f3f4d545cde21c0a3"
    }
    payload = {
        "input": text,
        "model": "text-embedding-3-small"
    }

    response = requests.post(url, headers=headers, json=payload)
    data = response.json()
    
    embedding = np.array(data['data'][0]['embedding'])
    
    return embedding
```
